```python
import pandas as pd
import numpy as np


```

```python
data_dir = '/Users/Karolina/Downloads/MyData-2/endsong_'
dfs = [] # an empty list to store the data frames

for i in range(25):
    data = pd.read_json(data_dir+str(i)+'.json') # read data frame from json file
    dfs.append(data) # append the data frame to the list

df = pd.concat(dfs, ignore_index=True) # concatenate all the data frames in the list.
df.head()
```

```python
df.info()
```

```python
df['endTime'] = pd.to_datetime(df.ts) 

# rounding up miliseconds Played to more readable formats
df['sPlayed'] = df['ms_played']/(1000)
df['mPlayed'] = df['sPlayed']/(60)
df['hPlayed'] = df['sPlayed']/(60*60)
```
